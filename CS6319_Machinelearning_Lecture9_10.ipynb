### A categorical case-study

```python
kr = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/chess/king-rook-vs-king-pawn/kr-vs-kp.data")
```

```python
kr
```

```python
# every column is categorical and has 3195 non-null values, so we need one-hot encode
kr.info()
```

```python
# show the data is almost balanced, so we use accuracy scores
kr['won'].value_counts()
```

```python
# split into train and test
from sklearn.model_selection import train_test_split

krtrain, krtest = train_test_split(kr, test_size=0.2)
traindata = krtrain.iloc[:,0:35]
trainlabs = krtrain.iloc[:,36]
testdata = krtest.iloc[:,0:35]
testlabs = krtest.iloc[:,36]
```

```python
# apply label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
trainlabse = le.fit_transform(trainlabs)
testlabse = le.transform(testlabs)
```

```python
# one-hot encoding all the train and test data
from sklearn.preprocessing import OneHotEncoder
catenc = OneHotEncoder(handle_unknown='ignore')
trainhot = catenc.fit_transform(traindata).toarray()
testhot = catenc.transform(testdata).toarray()
```

```python
# use shuffle-split cross-validation to test some classifiers
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC

split = ShuffleSplit(n_splits=10, test_size=0.2)
classifier = LinearSVC()
scores = cross_val_score(classifier, trainhot, trainlabse, scoring= "accuracy", cv=split)
print("LS acc mean =", scores.mean(), " stddev =", scores.std())
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
scores = cross_val_score(classifier, trainhot, trainlabse, scoring="accuracy", cv=split)
print("RF acc mean =", scores.mean(), " stddev =", scores.std())
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
scores = cross_val_score(classifier, trainhot, trainlabse, scoring="accuracy", cv=split)
print("KN acc mean =", scores.mean(), " stddev =", scores.std())
```

```python
# apply to the test data
from sklearn.metrics import mean_absolute_error

classifier = RandomForestClassifier()
classifier.fit(trainhot, trainlabse)
y1 = classifier.predict(testhot)
print("Test error =", mean_absolute_error(testlabse, y1))
```

### K-means

```python
from sklearn.cluster import KMeans

X = np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])
km= KMeans(n_clusters=2).fit(X)
print(km.labels_)
```

```python
print(km.cluster_centers_)
```

```python
labs = km.predict([[0,0],[12,3]])
print(labs)
```

```python
print(km.inertia_)
```

```python
from sklearn.metrics import silhouette_score
s = silhouette_score(X, km.labels_)
print(s)
```
